## 3. (Recommended) Install ImageJ for the watershed segmentation

The paper's reconstruction segments the implicit field into fragments with an ImageJ 3D watershed. This needs a JDK + `pyimagej`. **Required to reproduce the paper's exact numbers**; if you skip it, the pipeline falls back to direct marching cubes (still produces a figure, but not the paper's watershed result).

> The first run downloads Fiji (a few hundred MB), so ImageJ initialization can take a few minutes.

```python
!apt-get -qq install -y openjdk-11-jdk maven > /dev/null
!pip install -q pyimagej
import os
os.environ['JAVA_HOME'] = '/usr/lib/jvm/java-11-openjdk-amd64'
```
